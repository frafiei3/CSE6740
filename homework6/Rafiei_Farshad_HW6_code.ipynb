```python
data = pd.read_csv('spambase.data', header=None, names=col_names)
```

```python
data.head()
```

```python
# There is no missing value in the data
data.describe()
```

```python
# Split the data into train and test data
from sklearn.model_selection import train_test_split
X = data[col_names[:-1]]
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2020)
```

## CART

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
cart = DecisionTreeClassifier()
cart.fit(X_train, y_train)
y_pred = cart.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print('Total misclassification rate for decision tree classifier: {}'.format(1-accuracy_score(y_test, y_pred)))
```

```python
from sklearn.tree import plot_tree, export_graphviz
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(25, 15))
plot_tree(cart, max_depth=3, fontsize=18, feature_names=data.columns, class_names=['non-spam', 'spam'])
plt.show()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
error_forest = []
n_tree = []
error_tree = []
for num_tree in range(1,100,2):
    forest = RandomForestClassifier(n_estimators=num_tree)
    forest.fit(X_train, y_train)
    y_pred = forest.predict(X_test)
    n_tree.append(num_tree)
    error_forest.append(1 - accuracy_score(y_test, y_pred))
    error_tree.append(0.079261)
    

fig, ax = plt.subplots(figsize=(10, 10))
plt.plot(n_tree, error_forest, 'bo--', linewidth=2, markersize=12)
plt.plot(n_tree, error_tree, 'r', linewidth=2, markersize=12)
plt.ylabel('Total misclassification errror rate', fontsize=18)
plt.xlabel('# of trees', fontsize=18)
plt.legend(('Random Forest', 'Decision Tree'), loc='upper right', fontsize=24)
```

## One class SVM

```python
from sklearn.svm import OneClassSVM
nonSpam_X_train = X_train.loc[y_train==0]
y_pred = OneClassSVM(gamma=0.5).fit(nonSpam_X_train).predict(X_test)
y_pred[y_pred == -1] = 0
accuracy_score(y_test, 1-y_pred)
```
